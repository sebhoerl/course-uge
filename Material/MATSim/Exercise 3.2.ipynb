```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
df_trips = gpd.read_parquet("../Exercises/trips.parquet")
```

```python
df_trips
```

```python
def format_time(value):
    hour = int(value // 3600)
    minute = int((value % 3600) // 60)
    second = int(value % 60)
    
    return "{:02d}:{:02d}:{:02d}".format(hour, minute, second)

format_time(9 * 3600 + 300 + 2)
```

```python
content = []

content.append("<?xml version=\"1.0\" encoding=\"utf-8\"?>")
content.append("<!DOCTYPE population SYSTEM \"http://www.matsim.org/files/dtd/population_v6.dtd\">")
content.append("<population>")

for index, row in df_trips.iterrows():
    content.append("  <person id=\"{}\">".format(index))
    content.append("    <plan>".format(index))
    
    content.append("      <activity type=\"generic\" x=\"{}\" y=\"{}\" end_time=\"{}\" />".format(
          row["geometry"].coords[0][0], row["geometry"].coords[0][1], format_time(row["departure_time"])
    ))

    content.append("      <leg mode=\"car\" />")

    content.append("      <activity type=\"generic\" x=\"{}\" y=\"{}\" />".format(
          row["geometry"].coords[1][0], row["geometry"].coords[1][1])
    )
    
    content.append("    </plan>")
    content.append("  </person>")

content.append("</population>")
```

```python
content
```

```python
with open("plans.xml", "w+") as f:
    f.write("\n".join(content))
```
